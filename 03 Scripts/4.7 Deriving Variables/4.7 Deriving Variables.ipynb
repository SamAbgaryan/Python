# creating a function

```python
def price(row):
    if row["prices"]<=5:
        return "low-range"
    elif row["prices"]>5 and row["prices"]<=15:
        return "mid-range"
    elif row["prices"]>15:
        return "high-range"
    else:
        return "Not enough data"
    
```

```python
df.shape
```

# executing the function

```python
df["price_range"]=df.apply(price, axis=1)
```

## 2 two busiest and two slowest days of the week

```python
df_prods_ords_merge["order_dow"].value_counts(dropna=False)
```

```python
# creating a for loop to get the result needed
results=[]
for x in df_prods_ords_merge["order_dow"]:
    if x==0 or x==1:
        results.append("Busiest days")
    elif x==4 or x==3:
        results.append("Slowest days")
    else:
        results.append("Regular days")
```

# 4 Most orders, Average orders, and Fewest orders.

```python
# getting the count of the hours of the day
df_prods_ords_merge["order_hour_of_day"].value_counts(dropna=False)
```

```python
# creating a for loop to itereate 
hours=[]
for x in df_prods_ords_merge["order_hour_of_day"]:
    if x==10 or x==11 or x==14 or x==15 or x==13 or x==12 or x==16 or x==9:
        hours.append("Most orders")
    elif x==17 or x==8 or x==18 or x==19 or x==20 or x==7 or x==21 or x==22:
        hours.append("Average orders")
    else:
        hours.append("Fewest orders")
```

```python
hours
```

```python
# adding a column
df_prods_ords_merge["busiest_period_of_day"]=hours
```

```python
# checking value count
df_prods_ords_merge["busiest_period_of_day"].value_counts(dropna=False)
```
